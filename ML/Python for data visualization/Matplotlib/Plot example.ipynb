```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g-')  
plt.xlabel('X Lable')
plt.ylabel('Y Lable')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes1 = fig.add_axes([0.45,0.2,0.4,0.3])
axes.plot(x,y)
axes1.plot(y,x)
axes.set_title('LARGE PLOT')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()

axes[0].plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')
ax.plot(y,x)
ax.legend(loc=(0.5,0.1))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green')
```
